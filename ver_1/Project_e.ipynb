```python
import csv
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# CSV 파일 경로
csv_path = "archive/NFLX.csv"

# CSV 파일 읽어오기 (첫 번째, 마지막 열 제외)
df = pd.read_csv(csv_path, usecols=range(0, 5))

# DataFrame 출력
df
```

```python
df.dtypes
```

```python
df["Date"] = pd.to_datetime(df["Date"])

date_after_2022 = df[df['Date']>= "2022-01-01"]
date_after_2022
```

```python
plt.plot(date_after_2022['Date'], date_after_2022['Low'], label='Low')
plt.plot(date_after_2022['Date'], date_after_2022['High'], label='High')
plt.plot(date_after_2022['Date'], date_after_2022['Close'], label='Close')


# 그래프 제목 설정
plt.title('High, Low and Close Prices Since January 2022')

# x축 레이블 설정
plt.xlabel('Date')

# y축 레이블 설정
plt.ylabel('Price')

# 범례 표시
plt.legend()
plt.xticks(rotation=45)

# 그래프 표시
plt.show()
```
